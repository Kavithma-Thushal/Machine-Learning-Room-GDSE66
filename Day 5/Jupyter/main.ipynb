# Matplotlib

## The pyplot interface

```python
from matplotlib import pyplot as plt

plt.plot([1, 2, 3], [4, 5, 1])
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line plot')
plt.show()
```

### Basic Plot Types

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.show()
```
