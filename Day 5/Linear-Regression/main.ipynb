| Dosage (mg) | Blood Pressure (mmHg) |

### Step 1: Loading Data Set

```python
import pandas as pd

data = {
    'Dosage (mg)': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    'Blood Pressure (mmHg)': [120, 125, 132, 140, 145, 150, 158, 165, 170, 178]
}

df = pd.DataFrame(data)
print(df)
```

### Step 2: Visualizing Data Set

```python
import matplotlib.pyplot as plt

plt.scatter(df['Dosage (mg)'], df['Blood Pressure (mmHg)'], color='blue', label='Dosage')
plt.xlabel('Dosage (mg)')
plt.ylabel('Blood Pressure (mmHg)')
plt.title('Dosage vs. Blood Pressure')
plt.legend()
plt.show()
```

### Step 3: Preparing Data Set

```python
# Separate feature and target variables
X = df[['Dosage (mg)']].values  # Feature (Dosage)
y = df['Blood Pressure (mmHg)'].values

print(X)
print(y)
```

### Step 4: Define and Train the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
theta = model.intercept_
theta_1 = model.coef_[0]
print(f'Intercept (theta_0): {theta}')
print(f'Coefficient for Dosage (theta_1): {theta_1}')
```

### Step 5: Make Predictions

```python
# Predict the Blood Pressure for a specific Dosage
predicted_blood_pressure = model.predict([[2500]])
print(predicted_blood_pressure)

# Predict Blood Pressure for all Dosages
predicted_blood_pressure_all = model.predict(X)
print('Predicted Blood Pressure:', predicted_blood_pressure_all)
```

### Step 6: Evaluate the Model

```python
# Compare actual and predicted Blood Pressure values
comparison = pd.DataFrame({'Actual': y, 'Predicted': predicted_blood_pressure_all, 'Difference': y - predicted_blood_pressure_all})
print(comparison)
```
